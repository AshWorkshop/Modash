# 8. How do I apply multiple filter criteria to a pandas DataFrame?

```python
import pandas as pd
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
movies[(movies.duration >= 200) & (movies.genre == 'Drama')]  #use '&' but not 'and'
```

```python
movies.genre.isin(['Crime', 'Drama', 'Action']);
```

```python
movies[movies.genre.isin(['Crime', 'Drama', 'Action'])];
```

# 13. When should I use a "groupby" in pandas?

```python
import pandas as pd
```

```python
drinks = pd.read_csv('http://bit.ly/drinksbycountry')
```

```python
drinks.head()
```

```python
drinks.beer_servings.mean()
```

```python
drinks.groupby('continent').beer_servings.agg(['count', 'min', 'max', 'mean'])
```

```python
drinks.groupby('continent').mean()
```

```python
%matplotlib inline
```

```python
drinks.groupby('continent').mean().plot(kind='bar')
```
